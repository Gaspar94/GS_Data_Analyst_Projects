```python
# imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read datasets
# This dataset include 
df = pd.read_csv('Electric_Vehicle_Population_Data.csv')
```

```python
# General info about the columns in this dataset
df.columns
# df.head()
# df.describe()
```

```python
print(f"Number of rows : {len(df)}")
```

```python
df.drop_duplicates(inplace=True)
print(f"Number of rows : {len(df)}")
```

```python
df['Make'].value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of cars by make")
plt.ylabel('Number of cars')
plt.xlabel('Make')
```

```python
# Detecting posible outliers
sns.boxplot(x=df['Model Year'])
```

```python
cars_brand = df.groupby(['Make']).count()
cars_brand = cars_brand['Model'].sort_values(ascending=False)
cars_brand
```

```python
# create histogram
plt.hist(df['Model Year'], range=(2010,2024),bins=14,color='skyblue', edgecolor='black')
# Add title
plt.title('Distribution of cars by year')
# Add labels for axis
plt.xlabel('Year')
plt.ylabel('Number of cars')
# display histogram
plt.show()
```
